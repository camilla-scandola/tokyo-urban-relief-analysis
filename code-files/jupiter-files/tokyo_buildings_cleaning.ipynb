```python
import numpy as np
import pandas as pd
import ast
import numbers

def max_flood_depth(v):
    if v is None or (isinstance(v, float) and np.isnan(v)):
        return np.nan

    # If it's a string like "[1.84, 3.31, 13.98, 0.17]"
    if isinstance(v, str):
        try:
            v = ast.literal_eval(v)
        except Exception:
            return np.nan

    nums = []

    def walk(x):
        # numpy arrays, lists, tuples
        if isinstance(x, (list, tuple, np.ndarray)):
            for t in x:
                walk(t)
        # any numeric type (python, numpy, decimal, etc.)
        elif isinstance(x, numbers.Number) and not pd.isna(x):
            nums.append(float(x))

    walk(v)
    return max(nums) if nums else np.nan

df["max_flood_depth"] = df["value"].apply(max_flood_depth)
```
