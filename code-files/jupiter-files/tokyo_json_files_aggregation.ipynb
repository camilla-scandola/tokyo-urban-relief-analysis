```python
import geopandas as gpd
import pandas as pd
import glob
import os

folder = "/Users/camillascandola/Desktop/tokyo_json"
files = glob.glob(os.path.join(folder, "*.geo.json"))
print("Files found:", len(files))

gdfs = [gpd.read_file(f) for f in files]
tokyo_wards = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))

print(tokyo_wards.columns)

# (optional) ensure it's actually a string column
tokyo_wards["name"] = tokyo_wards["name"].astype(str)

tokyo_wards["name"] = (
    tokyo_wards["name"]
    .str.replace("-ku", "", regex=False)
    .str.strip()
)

output_path = "/Users/camillascandola/Desktop/tokyo_wards.geojson"
tokyo_wards.to_file(output_path, driver="GeoJSON")
print("Saved to:", output_path)
```
